```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players=len(purchase_data["SN"].unique())
pd.DataFrame([total_players], ["Total Players"])
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items=len(purchase_data["Item Name"].unique())
average_price=round(purchase_data["Price"].mean(),2)
total_purchases=purchase_data["Purchase ID"].count()
total_revenue=purchase_data["Price"].sum()


d={"Number of Unique Items":[unique_items], "Average Price":[average_price], "Number of Purchases":[total_purchases], "Total Revenue":[total_revenue]}
summary=pd.DataFrame(d)
summary["Average Price"]=summary["Average Price"].map("${:,.2f}".format)
summary["Total Revenue"]=summary["Total Revenue"].map("${:,.2f}".format)
summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
players=purchase_data.drop_duplicates(subset=["SN"])
male=players.loc[players["Gender"]=="Male"]
female=players.loc[players["Gender"]=="Female"]
other=players.loc[players["Gender"]=="Other / Non-Disclosed"]

female_count=female["SN"].count()
male_count=male["SN"].count()
other_count=other["SN"].count()

gender_demo=pd.DataFrame([male_count, female_count, other_count], index=["Male", "Female", "Other"], columns=["Total Players"])
gender_demo['Percentage']=gender_demo['Total Players']/total_players*100
gender_demo["Percentage"]=gender_demo["Percentage"].map("{0:.2f}%".format)
gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
relevant_info=purchase_data[["Gender", "Price"]]
gender_counts=relevant_info.groupby(['Gender']).count()
gender_totals=relevant_info.groupby(['Gender']).sum()
gender_means=relevant_info.groupby(['Gender']).mean()
x=[female_count, male_count, other_count]

purchasing_analysis_gender=pd.merge(gender_counts,gender_means, on="Gender")



purchasing_analysis_gender=pd.merge(purchasing_analysis_gender, gender_totals, on="Gender")
purchasing_analysis_gender['Total Players']=x
purchasing_analysis_gender=purchasing_analysis_gender.rename(columns={"Price_x":"Number of Purchases","Price_y" : "Average Purchase Price", "Price":"Total Purchase Value"})
purchasing_analysis_gender["Average Total Purchase"]=purchasing_analysis_gender["Total Purchase Value"]/purchasing_analysis_gender["Total Players"]
purchasing_analysis_gender["Total Purchase Value"]=purchasing_analysis_gender["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_gender["Average Total Purchase"]=purchasing_analysis_gender["Average Total Purchase"].map("${:,.2f}".format)
purchasing_analysis_gender["Average Purchase Price"]=purchasing_analysis_gender["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis_gender=purchasing_analysis_gender.drop(columns=["Total Players"])
purchasing_analysis_gender
```
